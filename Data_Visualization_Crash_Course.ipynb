## Excercise 1 : Relative Permeability Plots.

```python
Sw = np.linspace(0,1,100)

krw = Sw**3 ; kro = (1-Sw)**3
```

We can also customize our Plotting styles : 'default' , 'seaborn' etc.

```python
plt.style.use('default')
```

```python
#Step 1 : Plotting the first plot
plt.plot(Sw,krw)

#axis twinning

#Step 2 : Plotting the second plot
plt.plot(Sw,kro)
```

```python
#ADDING a GRID

#Step 1 : Plotting the first plot
plt.plot(Sw,krw,lw=5)

#Step 2 : Plotting the second plot
plt.plot(Sw,kro,lw=7)

#Step 3 : Plotting the Grid
plt.grid()


#plt.plot(x,y1,x,y2)
```

Customizing Our plot with more Details.

```python
#Labels and Legend. 
plt.plot(Sw,krw , label='Water Rel-Perm')
plt.plot(Sw,kro, label='Oil Rel-Perm')

plt.grid()
plt.legend()
```

```python
#Xlabel , Ylabel & Title
plt.plot(Sw,krw , lw = 3,label='Krw')
plt.plot(Sw,kro, lw = 3, label='Kro')

plt.grid()

#Xlabel,Ylabel
plt.xlabel('Sw(Fraction)')
plt.ylabel('Krw & Kro(Fraction)')
plt.legend()

#title
plt.title('Rel-Perm Plots')

```

```python
#SUBPLOTS.

#Step 1 : Data Creation
x = np.linspace(0,25)
y1 = x**2 ; y2 = x**3

#Step 2 : Subplots. 
fig , ax = plt.subplots(nrows=1, ncols=2)
fig.set_figwidth(12)


#Plot 1
ax[0].plot(x,y1,'r*-')
ax[0].set_title('Power = 2')
ax[0].grid()


#Plot 2
ax[1].plot(x,y2,'k--')
ax[1].set_title('Power = 3')
ax[1].grid()
```

```python
#Step 1 : Data Creation
x = np.linspace(0,25)
y1 = x**2 ; y2 = x**3
y3 = x**4 ; y4 = x**5

#Step 2 : Subplots. 
fig , ax = plt.subplots(nrows=2, ncols=2)
fig.set_figwidth(12) ; fig.set_figheight(8)


#Plot 1
ax[0,0].plot(x,y1,'r*-')
ax[0,0].set_title('Power = 2')
ax[0,0].grid()


#Plot 2
ax[0,1].plot(x,y2,'k--')
ax[0,1].set_title('Power = 3')
ax[0,1].grid()


#Plot 3
ax[1,0].plot(x,y3,'bv-')
ax[1,0].set_title('Power = 4')
ax[1,0].grid()

#Plot 4
ax[1,1].plot(x,y4,'g*-')
ax[1,1].set_title('Power = 5')
ax[1,1].grid()
```

What if we want to make some inferences about the statistical nature of the 'Total-Bill' that customers pay.

## Example : DCA : Exponential Decline.

```python
t = np.linspace(1,1000)

dates = pd.date_range(start='1/10/2005', end='1/4/2021',periods=len(t))

d = dates.year


Q0 = 2000 #bbls/d
b = 0.05

Q = Q0*np.exp(-b*t)
```

```python
df_exp = pd.DataFrame({'t':t, 'Q':Q}, index=dates)
df_exp['year'] = d

df_exp.head()
```

```python
fig = px.line(data_frame = df_exp , x='t' , y='Q')
```

```python
fig.show()
```

# Final Excercise : Radial Flow Profile Example. Impact of Oil Viscosity

```python
rw = 0.5 #ft 
re = 2000 #ft

r = np.linspace(rw,re,1000)

Pe = 5000 

mu1 = 0.5 #cp 
mu2 = 10 #cp 

k = 1000 #mD 
h = 50 #ft

q = 1000 #bbl/d

B = 1

m = (141.2*q*mu1*B/k/h)

Pr1 = Pe - m*(np.log(re/r))


plt.plot(r,Pr1)
plt.xlabel('r(ft') ; plt.ylabel('P(r), psi')

Pwf1 = Pr1[0]

plt.axhline(Pwf1 , c='red')
plt.xlim(rw,re) ; plt.ylim(Pwf1-20, Pe)
```
